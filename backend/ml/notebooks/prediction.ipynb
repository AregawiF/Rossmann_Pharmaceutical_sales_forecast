```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

```

```python
df = pd.read_csv('../src/data/cleaned_data/cleaned_train_df.csv')
```

```python
df.head()
```

```python
print(df['StateHoliday'].unique())
print(df['HolidayPeriod'].unique())
```

```python
# Convert StateHoliday to numeric
state_holiday_mapping = {'0': 0, 'a': 1, 'b': 2, 'c': 3}
df['StateHoliday'] = df['StateHoliday'].replace(state_holiday_mapping).astype(int)

# Handle missing values in HolidayPeriod
df['HolidayPeriod'] = df['HolidayPeriod'].fillna("Not during holiday")

# Extract weekday/weekend feature
df['IsWeekend'] = df['DayOfWeek'].apply(lambda x: 1 if x >= 6 else 0)

# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Beginning, Mid, End of Month
df['MonthPart'] = df['Date'].dt.day.apply(lambda x: "Beginning" if x <= 10 else ("Mid" if x <= 20 else "End"))

# Example for Days to Holidays and Days after Holidays
# (Requires a holiday dataset with dates for accurate implementation)
holidays = pd.to_datetime(["2015-07-01", "2015-08-15"])  # Example holidays
df['DaysToHoliday'] = df['Date'].apply(lambda x: (holidays[holidays > x] - x).min().days if any(holidays > x) else -1)
df['DaysAfterHoliday'] = df['Date'].apply(lambda x: (x - holidays[holidays < x]).min().days if any(holidays < x) else -1)

# Preview the processed DataFrame
df.head()
```
